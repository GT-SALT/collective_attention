# Get sample posts for annotation
In which we collect some sample annotations (N=20) for someone else to try annotating.

```python
import codecs
import re
import numpy as np
import pandas as pd
```

## Load data

```python
annotated_text_file = '../../data/facebook-maria/all_group_sample_statuses_annotated.txt'
annotated_text_df = pd.read_csv(annotated_text_file, sep='\t', index_col=False)
annotated_text_df.columns = ['id', 'text']
annotated_text = annotated_text_df.loc[:, 'text']
print('\n'.join(annotated_text[:10]))
```

```python
topo_text_file = '../../data/facebook-maria/all_group_sample_statuses_annotated_topo.txt'
topo_text = [l.strip() for l in codecs.open(topo_text_file, 'r', encoding='utf-8')]
print('\n'.join(topo_text))
```

## Clean text
Remove brackets from `annotated\_text` and ID/lat-lon coordinates from `topo\_text`.

```python
bracket_matcher = re.compile('\[\[[^\]]+\]\]\{[A-Z/]+\}')
topo_matcher = re.compile('(?<=\[\[)([^\]]+)(?=\]\]\{[A-Z/]+\})')
test_str = 'Alguien tiene información de [[Río Abajo]]{CITY/TOPO} y [[Vista Monte]]{TOPO} . Por favor .'
for bracket_str, topo_str in zip(bracket_matcher.findall(test_str), topo_matcher.findall(test_str)):
    test_str = test_str.replace(bracket_str, topo_str)
print(test_str)
```

```python
bracket_matcher = re.compile('\[\[[^\]]+\]\]\{[A-Z/]+\}')
topo_matcher = re.compile('(?<=\[\[)([^\]]+)(?=\]\]\{[A-Z/]+\})')
def remove_brackets(s):
    bracket_txt = bracket_matcher.findall(s)
    topo_txt = topo_matcher.findall(s)
    for b, t in zip(bracket_txt, topo_txt):
        s = s.replace(b,t)
    return s
```

```python
np.random.seed(345)
sample_size = 20
sample_idx = np.random.choice(annotated_text_df.index, size=sample_size, replace=False)
annotated_text_sample = annotated_text_df.loc[sample_idx, :]
annotated_text_sample_txt = annotated_text_sample.loc[:, 'text'].values.tolist()
annotated_text_sample_with_topo = filter(lambda x: len(bracket_matcher.findall(x)) > 0, annotated_text_sample_txt)
print('%d posts with at least one toponym'%(len(annotated_text_sample_with_topo)))
annotated_text_sample_txt = map(remove_brackets, annotated_text_sample_txt)
annotated_text_sample.loc[:, 'text'] = annotated_text_sample_txt
print('\n'.join(annotated_text_sample.loc[:, 'text'][:5]))
```

Now write status and status number to file.

```python
# write text
annotated_text_sample_file_name = '../../data/facebook-maria/all_group_sample_statuses_blank_sample.txt'
with codecs.open(annotated_text_sample_file_name, 'w', encoding='utf-8') as annotated_text_sample_file:
    annotated_text_sample_file.write('\n'.join(annotated_text_sample.loc[:, 'text'].values.tolist()).decode('utf-8'))
# write IDs
annotated_text_sample_id_file_name = '../../data/facebook-maria/all_group_sample_statuses_blank_sample_ids.txt'
with codecs.open(annotated_text_sample_id_file_name, 'w', encoding='utf-8') as annotated_text_sample_id_file:
    annotated_text_sample_id_file.write('\n'.join(annotated_text_sample.loc[:, 'id'].values.tolist()))
```

Same thing! But with the toponym details removed.

```python
full_bracket_matcher = re.compile('\[\[[^\]]+\]\]\{[A-Z/]+\}\{[^\}]+\}')
full_topo_matcher = re.compile('\[\[[^\]]+\]\]\{[A-Z/]+\}(?=\{[^\}]+\})')
test_str = 'Hola DLB ... alguien tiene información de [[Valle Alto]]{TOPO}{4568505,18.01247,-66.01989} o [[Marin]]{CITY} ?'
for b, t in zip(full_bracket_matcher.findall(test_str), full_topo_matcher.findall(test_str)):
    test_str = test_str.replace(b, t)
print(test_str)
```

```python
full_bracket_matcher = re.compile('\[\[[^\]]+\]\]\{[A-Z/]+\}\{[^\}]+\}')
full_topo_data_matcher = re.compile('\[\[[^\]]+\]\]\{[A-Z/]+\}(?=\{[^\}]+\})')
def remove_full_brackets_and_extras(s):
    bracket_txt = full_bracket_matcher.findall(s)
    topo_txt = full_topo_matcher.findall(s)
    for b, t in zip(bracket_txt, topo_txt):
        s = s.replace(b, t)
    # also get rid of CITY/UNK crap
    s = s.replace('CITY', 'TOPO').replace('UNK', 'TOPO')
    # add extra bracket for annotation
    s = s.replace('{TOPO}', '{TOPO}{}')
    return s
```

```python
topo_text_clean = map(remove_full_brackets_and_extras, topo_text)
print('\n'.join(topo_text_clean))
```

```python
# write to file
topo_text_clean_file_name = '../../data/facebook-maria/all_group_sample_statuses_annotated_topo_blank.txt'
with codecs.open(topo_text_clean_file_name, 'w', encoding='utf-8') as topo_text_clean_file:
    topo_text_clean_file.write('\n'.join(topo_text_clean))
```
