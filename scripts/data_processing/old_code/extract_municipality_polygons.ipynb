## Extract polygons

```python
municipality_polygons = [(p['properties']['name'], p['geometry']['coordinates']) for p in place_list if p['properties']['name'] in municipalities]
print(municipality_polygons)
```

Wait a minute!! These aren't municipalities, these are towns. We need the county boundaries!

Downloaded from here: https://www.census.gov/geo/maps-data/data/cbf/cbf_counties.html

```python
county_shapes = fiona.open('../../data/geo_files/county_shape_files/cb_2016_us_county_500k.shp')
county_shape_list = list(county_shapes)
print(json.dumps(county_shape_list[:10], indent=2))
```

```python
PR_state_FP = "72"
PR_county_shape_list = filter(lambda x: x['properties']['STATEFP'] == PR_state_FP, county_shape_list)
print(len(PR_county_shape_list))
print(json.dumps(PR_county_shape_list, indent=2))
```

Collect the actual polygons.

```python
from shapely.geometry import shape, Point
municipality_geoms = [(c['properties']['NAME'], c['geometry']) for c in PR_county_shape_list]
municipality_shapes = [(n, shape(g)) for n,g in municipality_geoms]
```

## Containment tests

For every entity in the previously-collected list, determine which municipality (if any) it belongs to.

```python
entity_info = pd.read_csv('../../data/geo_files/PR_OSM/shp_files/combined_shp_data.tsv', sep='\t', index_col=False)
print(entity_info.head())
```

```python
def containment_test(lat_lon, municipality_shapes):
    point = Point(lat_lon)
    for municipality_name, municipality_shape in municipality_shapes:
        if(municipality_shape.contains(point)):
            return municipality_name
    return None
```

```python
test_location = entity_info[entity_info.loc[:, 'name'] == 'Guayama'].loc[:, ['lat', 'lon']].values[0]
print(test_location)
test_municipality = containment_test(test_location, municipality_shapes)
print(test_municipality)
```

It works! I think.

```python
print(entity_info.head().loc[:, ['lat', 'lon']].apply(lambda x: x.values.tolist()))
```

```python
entity_municipality = entity_info.loc[:, ['lat', 'lon']].apply(lambda p: containment_test(p.values.tolist(), municipality_shapes), axis=1)
```

```python
print(entity_municipality.head())
```

Join with entity info and rewrite to file!!

```python
entity_info.loc[:, 'municipality'] = entity_municipality
print(entity_info.head())
```

What is the distribution over all municipalities?

```python
municipality_counts = entity_municipality.value_counts()
print(municipality_counts)
```

Makes sense! Most documented entities are in the capital San Juan.
